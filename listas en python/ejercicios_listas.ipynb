## Ejercicio 3-4: Lista de Invitados a Cena

**Descripción:**
Si pudiese invitar a cualquiera, vivo o muerto, a cenar, ¿a quién invitaría? Haga una lista de al menos tres personas a las que le gustaría invitar a una cena y úsela para imprimir un mensaje para cada persona invitándole a cenar.

**Concepto:** Crear una lista con nombres y usar un bucle `for` para iterar sobre ella, imprimiendo un mensaje personalizado para cada invitado usando el método `.title()` para capitalizar los nombres.

```python
invitaciones = ['nelly','viviana','luis','camilo','mariana','mariajose','ines']
for invitacion in invitaciones:
    print(f'Te invito a una cena de navidad {invitacion.title()} para el día 24 de diciembre')
```

## Ejercicio 3-5: Cambiar la Lista de Invitados

**Descripción:**
Acaba de enterarse de que uno de sus invitados no podrá asistir, así que tiene que enviar un nuevo juego de invitaciones. Tendrá que pensar en otra persona para invitar.

**Requisitos:**
- Empiece con el programa del ejercicio 3-4. Añada una llamada a `print()` al final del programa indicando el nombre del invitado que no puede asistir.
- Modifique la lista sustituyendo el nombre del invitado que no puede venir por el de otra persona a la que va a invitar en su lugar.
- Imprima un segundo grupo de mensajes de invitación, uno para cada persona que haya en la lista ahora.

**Concepto:** Usar el método `.pop()` para remover un elemento, `.insert()` para agregar un elemento en una posición específica, y condicionales para personalizar mensajes.

```python
no_asistira = invitaciones.pop(3).title()
print(f'No podrá asistir a la reunión {no_asistira}')
invitaciones.insert(3, 'sofia')
print('Invitados a la fiesta de navidad')
print('\n'.join(invitaciones).title())
print('Invitaciones a la fiesta de navidad')
for invitacion2 in invitaciones:
    if invitacion2 == 'sofia':
        print(f'Has confirmado que asistirás {invitacion2.title()} eres bienvenida a la fiesta de navidad')
    else:
        print(f'Bienvenidos a la fiesta de navidad {invitacion2.title()}')
```

## Ejercicio 3-6: Más Invitados con una Mesa Más Grande

**Descripción:**
Acaba de encontrar una mesa más grande, así que dispone de más espacio. Piense en otros tres invitados más.

**Requisitos:**
- Empiece con el programa del ejercicio 3-4 o 3-5. Añada una llamada a `print()` al final para informar a la gente de que ha encontrado una mesa de comedor más grande.
- Use `insert()` para añadir un nuevo invitado al principio de la lista.
- Use `insert()` para añadir un nuevo invitado en mitad de la lista.
- Use `append()` para añadir un nuevo invitado al final de la lista.
- Imprima un nuevo juego de mensajes de invitación, uno para cada persona de la lista.

**Concepto:** Consolidar el uso de diferentes métodos para modificar listas: `.insert()` para posiciones específicas y `.append()` para agregar al final.

```python
# Reinicializar la lista de invitaciones para este ejercicio
invitaciones = ['nelly','viviana','luis','sofia','mariana','mariajose','ines']

print('¡He encontrado una mesa más grande!\n')

# Agregar un invitado al principio de la lista
invitaciones.insert(0, 'carlos')

# Agregar un invitado en mitad de la lista
invitaciones.insert(len(invitaciones) // 2, 'diana')

# Agregar un invitado al final de la lista
invitaciones.append('roberto')

print('Nuevos invitados a la fiesta de navidad:')
for i, invitacion in enumerate(invitaciones, 1):
    print(f'{i}. {invitacion.title()}')

print('\n--- Nuevas invitaciones ---\n')
for invitacion in invitaciones:
    print(f'Te invitamos a la fiesta de navidad {invitacion.title()}. ¡Esperamos tu confirmación!')
```

## 3-7. Reducir la lista de invitados: 

**Descripción:**
Acaba de descubrir que la mesa no llegará a tiempo para la cena y solo tiene espacio para dos invitados.

**Requisitos:**
- Empiece con el programa de 3-6. Añada una línea que imprima un mensaje diciendo que solo puede invitar a dos personas a cenar.
- Use `pop()` para eliminar invitados de la lista de uno en uno hasta que solo queden dos. Cada vez que retire un nombre de la lista, imprima un mensaje para esa persona diciendo que lo siente, pero que no puede invitarle a cenar.
- Imprima un mensaje para cada una de las dos personas que quedan en la lista informándoles de que siguen invitados.
- Use `del` para borrar los dos últimos nombres de la lista y que se quede vacía. Imprima la lista para asegurarse de que realmente tiene una lista vacía alfinal del programa.

```python
print('Le informo que solo podré invitar a 2 personas porque la mesa no llegará a tiempo \n')
invitaciones = ['nelly','viviana','luis','sofia','mariana','mariajose','ines']
# Quise hacer este ejercicio con un bucle for para iterar en cado uno de los nombres pero al Iterar sobre una lista mientras la modifico causa que se salten elementos y el metodo pop() comiensa a eliminar nobres des de el ultimo si no se define una posicion
# Cuando invitaciones.pop() se ejecuta, elimina el ÚLTIMO elemento, no el actual 
for eliminar in invitaciones:
    if eliminar != 'nelly' and eliminar != 'luis':
        eliminados = invitaciones.pop()
        print(f'Lo siento, pero por motivos de que me informaron que la mesa no llegará a tiempo, no podré invitarte a la fiesta de navidad {eliminados.title()}')
    else:
        print(f'Estás invitado a la fiesta de navidad {eliminar.title()}')

# Aquí elimino a viviana y a sofia
print(f'Lo siento, pero por motivos de que me informaron que la mesa no llegará a tiempo, no podré invitarte a la fiesta de navidad {invitaciones.pop(1).title()}')
print(f'Lo siento, pero por motivos de que me informaron que la mesa no llegará a tiempo, no podré invitarte a la fiesta de navidad {invitaciones.pop(2).title()}\n')

print(invitaciones)
del invitaciones[:]
print(f'La lista de invitados está vacía: {invitaciones}')
```
